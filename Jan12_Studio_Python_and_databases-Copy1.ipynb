With your dataframe at the ready, create a new database called `tv`. Add a new table to your database called `shows` using the data in the dataframe. 

```python
#how to create a new database
con = sl.connect('tv.db')
print ("Opened database successfully")
```

```python
#Add a new table to your database called shows
with con:
     con.execute( """DROP TABLE shows""")
#created, then dropped the table - needed new name
```

With your new table and database set up, print out the top 20 records in the `shows` table.

```python
#turn db into table
data.to_sql('shows', con)
```

```python
with con:
    data = con.execute('SELECT * FROM shows where "id" between 1 and 20')
    for row in data:
        print(row)
#id is a number, we used this to pull up 20 rows
```
